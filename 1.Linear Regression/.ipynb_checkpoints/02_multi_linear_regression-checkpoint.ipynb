# 多元线性回归

这个 notebook 展示最基础的多元线性回归：使用多个特征 $x_1, x_2, \ldots, x_n$ 来预测标签 $y$。为了便于说明，下面的示例统一使用两个特征 $x_1$ 和 $x_2$。

## 1. 数据准备

设每个样本有两个特征：$x_1$ 和 $x_2$，标签为 $y$。下面给出 10 行样本数据：

| 编号 | $x_1$ | $x_2$ | $y$ |
| --- | --- | --- | --- |
| 1 | 1 | 2 | 8 |
| 2 | 2 | 1 | 9 |
| 3 | 3 | 0 | 10 |
| 4 | 0 | 4 | 9 |
| 5 | 4 | 3 | 19 |
| 6 | 5 | 2 | 20 |
| 7 | 2 | 5 | 17 |
| 8 | 6 | 1 | 21 |
| 9 | 3 | 4 | 18 |
| 10 | 7 | 2 | 26 |

这组数据满足下面这条线性关系：

$$
y = 3x_1 + 2x_2 + 1
$$


```python
X = [
    [1, 2],
    [2, 1],
    [3, 0],
    [0, 4],
    [4, 3],
    [5, 2],
    [2, 5],
    [6, 1],
    [3, 4],
    [7, 2],
]

y = [8, 9, 10, 9, 19, 20, 17, 21, 18, 26]

list(zip(X, y))
```

## 2. 多元线性回归公式

多元线性回归模型可以写成：

$$
\hat{y} = w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b
$$

也可以写成向量形式：

$$
\hat{y} = \vec{w} \cdot \vec{x} + b
$$

如果写到第 $i$ 个样本，则有：

$$
\hat{y}^{(i)} = \vec{w} \cdot \vec{x}^{(i)} + b
$$

其中：

- $\vec{x}^{(i)} = [x_1^{(i)}, x_2^{(i)}, \ldots, x_n^{(i)}]$ 表示第 $i$ 个样本的特征向量
- $\vec{w} = [w_1, w_2, \ldots, w_n]$ 表示各个特征对应的权重
- $b$ 表示偏置
- $\hat{y}^{(i)}$ 表示模型对第 $i$ 个样本的预测值


## 3. 向量与向量点积

在多元线性回归中，一个样本通常不再只有一个特征，而是由多个特征组成。此时，把这些特征放在一起，就可以写成一个向量。

例如，当一个样本有两个特征 $x_1$ 和 $x_2$ 时，可以把它写成：

$$
\vec{x} = [x_1, x_2]
$$

对应地，模型参数也可以写成一个向量：

$$
\vec{w} = [w_1, w_2]
$$

如果有更多特征，那么同样可以写成：

$$
\vec{x} = [x_1, x_2, \ldots, x_n], \quad \vec{w} = [w_1, w_2, \ldots, w_n]
$$

所谓“向量点积”，就是把两个向量中对应位置的元素分别相乘，再把结果相加。

例如：

$$
\vec{w} \cdot \vec{x} = w_1 x_1 + w_2 x_2
$$

更一般地，有：

$$
\vec{w} \cdot \vec{x} = w_1 x_1 + w_2 x_2 + \cdots + w_n x_n
$$

所以，多元线性回归写成向量形式：

$$
\hat{y} = \vec{w} \cdot \vec{x} + b
$$

本质上只是把原来逐项展开的式子：

$$
\hat{y} = w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b
$$

压缩成了更简洁的写法。


## 4. 代价函数（Cost Function）

在多元线性回归中，常用均方误差形式的代价函数：

$$
J(\vec{w}, b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2
$$

把预测公式代入后可得：

$$
J(\vec{w}, b) = \frac{1}{2m} \sum_{i=1}^{m} (\vec{w} \cdot \vec{x}^{(i)} + b - y^{(i)})^2
$$

其中 $m$ 表示样本数量。

## 5. 更新参数的公式

对每一个权重 $w_j$，梯度为：

$$
\frac{\partial J(\vec{w}, b)}{\partial w_j} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) x_j^{(i)}
$$

对偏置 $b$，梯度为：

$$
\frac{\partial J(\vec{w}, b)}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})
$$

因此，参数更新公式为：

$$
w_j := w_j - \alpha \frac{\partial J(\vec{w}, b)}{\partial w_j}, \quad j = 1, 2, \ldots, n
$$

$$
b := b - \alpha \frac{\partial J(\vec{w}, b)}{\partial b}
$$

其中 $\alpha$ 是学习率。

## 6. 训练过程示例

下面用这 10 个样本说明一次梯度下降是如何更新参数的。为了便于说明，先设初始参数为：

$$
w_1 = 0, \quad w_2 = 0, \quad b = 0
$$

学习率取为：

$$
\alpha = 0.01
$$

此时模型预测公式为：

$$
\hat{y} = 0 \cdot x_1 + 0 \cdot x_2 + 0 = 0
$$

也就是说，当前所有样本的预测值都为 0。

***第一步：计算代价函数***

共有 $m = 10$ 个样本，因此：

$$
J(\vec{w}, b) = \frac{1}{2 \times 10} \sum_{i=1}^{10} (\hat{y}^{(i)} - y^{(i)})^2
$$

由于当前所有 $\hat{y}^{(i)} = 0$，所以：

$$
J(0,0,0) = \frac{1}{20} (8^2 + 9^2 + 10^2 + 9^2 + 19^2 + 20^2 + 17^2 + 21^2 + 18^2 + 26^2)
$$

$$
= \frac{2817}{20} = 140.85
$$

***第二步：计算 $w_1$ 的梯度***

$$
\frac{\partial J}{\partial w_1} = \frac{1}{10} \sum_{i=1}^{10} (\hat{y}^{(i)} - y^{(i)}) x_1^{(i)}
$$

由于 $\hat{y}^{(i)} = 0$，代入数据可得：

$$
\frac{\partial J}{\partial w_1} = \frac{1}{10} [(-8)\cdot 1 + (-9)\cdot 2 + (-10)\cdot 3 + (-9)\cdot 0 + (-19)\cdot 4 + (-20)\cdot 5 + (-17)\cdot 2 + (-21)\cdot 6 + (-18)\cdot 3 + (-26)\cdot 7]
$$

$$
= \frac{-628}{10} = -62.8
$$

***第三步：计算 $w_2$ 的梯度***

$$
\frac{\partial J}{\partial w_2} = \frac{1}{10} \sum_{i=1}^{10} (\hat{y}^{(i)} - y^{(i)}) x_2^{(i)}
$$

代入数据可得：

$$
\frac{\partial J}{\partial w_2} = \frac{1}{10} [(-8)\cdot 2 + (-9)\cdot 1 + (-10)\cdot 0 + (-9)\cdot 4 + (-19)\cdot 3 + (-20)\cdot 2 + (-17)\cdot 5 + (-21)\cdot 1 + (-18)\cdot 4 + (-26)\cdot 2]
$$

$$
= \frac{-388}{10} = -38.8
$$

***第四步：计算 $b$ 的梯度***

$$
\frac{\partial J}{\partial b} = \frac{1}{10} \sum_{i=1}^{10} (\hat{y}^{(i)} - y^{(i)})
$$

$$
= \frac{-(8 + 9 + 10 + 9 + 19 + 20 + 17 + 21 + 18 + 26)}{10} = \frac{-157}{10} = -15.7
$$

***第五步：更新参数***

$$
w_1 := 0 - 0.01 \times (-62.8) = 0.628
$$

$$
w_2 := 0 - 0.01 \times (-38.8) = 0.388
$$

$$
b := 0 - 0.01 \times (-15.7) = 0.157
$$

所以，一次更新后参数从：

$$
(w_1, w_2, b) = (0, 0, 0)
$$

变成：

$$
(w_1, w_2, b) = (0.628, 0.388, 0.157)
$$


## 7. 如何判断收敛

常见的一种方式，是比较这一轮和上一轮的代价函数变化是否已经足够小：

$$
\left| J_{\text{prev}} - J(\vec{w}, b) \right| < \epsilon
$$

其中：

- $J_{\text{prev}}$ 表示上一轮的代价函数值
- $J(\vec{w}, b)$ 表示当前这一轮的代价函数值
- $\epsilon$ 表示收敛阈值，是一个很小的正数

如果满足这个条件，就可以认为模型已经基本收敛。

## 8. 训练过程总结

多元线性回归的训练目标，就是不断调整参数向量 $\vec{w}$ 和偏置 $b$，让代价函数 $J(\vec{w}, b)$ 尽可能小。

| 符号 | 释义 |
| --- | --- |
| $$\{(\vec{x}^{(1)}, y^{(1)}), (\vec{x}^{(2)}, y^{(2)}), \ldots, (\vec{x}^{(m)}, y^{(m)})\}$$ | 训练数据 |
| $\vec{x}$ | 特征向量 |
| $y$ | 真实值 |
| $\hat{y}$ | 预测值 |
| $\vec{w}$ | 权重向量 |
| $b$ | 偏置 |
| $\alpha$ | 学习率 |
| $\text{num\_iters}$ | 最大迭代次数 |
| $\epsilon$ | 收敛阈值 |
| $J(\vec{w}, b)$ | 代价函数 |
| $J_{\text{prev}}$ | 上一轮的代价函数值 |

***初始化：***

$$
\vec{w} = [0, 0, \ldots, 0], \quad b = 0, \quad J_{\text{prev}} = +\infty
$$

***for (i=0, i<num_iters, i++):***

***{***

$$
\begin{aligned}
&\text{1. 计算当前代价函数：其中 } \hat{y}^{(i)} = \vec{w} \cdot \vec{x}^{(i)} + b\\
&J(\vec{w}, b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2 = \frac{1}{2m} \sum_{i=1}^{m} (\vec{w} \cdot \vec{x}^{(i)} + b - y^{(i)})^2\\[6pt]
&\text{2. 判断是否收敛：}\\
&\text{若 } \left| J_{\text{prev}} - J(\vec{w}, b) \right| < \epsilon \text{，则停止训练；否则令 } J_{\text{prev}} = J(\vec{w}, b)\\[6pt]
&\text{3. 若未收敛，则更新参数：}\\
&w_j = w_j - \alpha \frac{\partial J(\vec{w}, b)}{\partial w_j} = w_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) x_j^{(i)}, \quad j = 1, 2, \ldots, n\\[6pt]
&b = b - \alpha \frac{\partial J(\vec{w}, b)}{\partial b} = b - \alpha \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})\\
\end{aligned}
$$

***}***

最终输出训练后的参数向量 $\vec{w}$ 和偏置 $b$。

## 9. 示例代码

下面给出一个完整的多元线性回归训练示例，包含代价函数、梯度下降、模型训练和预测过程。

```python
# 代价函数：计算当前参数下的平均平方误差的一半
def cost_function(X, y, w, b):
    m = len(X)
    total_cost = 0.0

    for i in range(m):
        # 计算第 i 个样本的预测值
        y_hat = sum(w[j] * X[i][j] for j in range(len(w))) + b
        total_cost += (y_hat - y[i]) ** 2

    return total_cost / (2 * m)
```

```python
# 梯度下降：根据所有样本的梯度更新一次参数
def gradient_descent(X, y, w, b, alpha):
    m = len(X)
    n = len(w)
    dj_dw = [0.0] * n
    dj_db = 0.0

    for i in range(m):
        # 先计算预测误差，再分别累加到每个参数的梯度上
        y_hat = sum(w[j] * X[i][j] for j in range(n)) + b
        error = y_hat - y[i]

        for j in range(n):
            dj_dw[j] += error * X[i][j]

        dj_db += error

    dj_dw = [grad / m for grad in dj_dw]
    dj_db = dj_db / m

    for j in range(n):
        w[j] = w[j] - alpha * dj_dw[j]

    b = b - alpha * dj_db

    return w, b
```

```python
# 模型训练：先计算代价函数，再判断收敛，未收敛时再更新参数
def train_model(X, y, alpha=0.01, num_iters=10000, epsilon=1e-9):
    n = len(X[0])
    w = [0.0] * n
    b = 0.0
    j_prev = float('inf')
    history = []
    last_iter = 0
    last_cost_diff = float('inf')
    prev_log = None
    current_log = None

    def log_iteration(iter_idx, w, b, current_cost, cost_diff):
        # 输出当前迭代的参数、代价函数和收敛阈值信息
        print(
            f"[{iter_idx:>4}/{num_iters}]: w = {[round(value, 6) for value in w]}, "
            f"b = {b:.6f}, cost = {current_cost:.12f}, "
            f"j_prev - current_cost = {cost_diff}, epsilon = {epsilon}"
        )

    for i in range(num_iters):
        # 1. 计算当前代价函数值
        current_cost = cost_function(X, y, w, b)
        history.append(current_cost)

        cost_diff = j_prev - current_cost if j_prev != float('inf') else float('inf')
        last_iter = i
        last_cost_diff = cost_diff
        current_log = (i, w[:], b, current_cost, cost_diff)

        if i % 100 == 0:
            log_iteration(*current_log)

        # 2. 判断是否收敛
        if abs(j_prev - current_cost) < epsilon:
            if prev_log is not None and prev_log[0] == i - 1 and (i - 1) % 100 != 0:
                log_iteration(*prev_log)
            break

        j_prev = current_cost

        # 3. 若未收敛，则使用梯度下降更新参数
        w, b = gradient_descent(X, y, w, b, alpha)
        prev_log = current_log

    log_iteration(last_iter, w, b, history[-1], last_cost_diff)

    return w, b, history
```

```python
# 使用前面定义的函数训练模型并做预测
X = [
    [1, 2],
    [2, 1],
    [3, 0],
    [0, 4],
    [4, 3],
    [5, 2],
    [2, 5],
    [6, 1],
    [3, 4],
    [7, 2],
]

y = [8, 9, 10, 9, 19, 20, 17, 21, 18, 26]

w, b, history = train_model(X, y, alpha=0.01, num_iters=20000, epsilon=1e-9)

print(f"【训练完成】: w = {[round(value, 6) for value in w]}, b = {b:.6f}, cost = {history[-1]:.12f}")

# 使用训练后的模型作预测
X_test = [[2, 3], [4, 1], [6, 5]]
predictions = [sum(w[j] * sample[j] for j in range(len(w))) + b for sample in X_test]

for sample, pred in zip(X_test, predictions):
    print(f"x = {sample}, 【预测值】 = {pred:.6f}")
```
